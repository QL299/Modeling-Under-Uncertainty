```python
result = []
for c in range (30, 50):
    mc = np.zeros((21,21))
    for i in range(21):
        for j in range(21):
            if j == i:
                mc[i][j] = (c-30)/20
            if i == j - 1:
                mc[i][j] = 1 - (c-30)/20
            if i == j & i == 20:
                mc[i][j] = 1
    R = np.full(shape = 21, fill_value = (1 - (c-30)/20)*c, dtype = np.float)
    L = np.identity(21)
    beta = 0.955
    V = np.linalg.inv(L - beta*mc) @ R
    for i in range(21):
        r = []
        r.append(c)
        r.append(i)
        r.append(V[20 - i]- 20*i)
        result.append(r)
print(result)
```
